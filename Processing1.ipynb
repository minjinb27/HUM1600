```python
# Problem 2
user_x = int(input("Please enter integer for x: "))
user_y = int(input("Please enter integer for y: "))
if user_x > user_y:
  print("X is larger than Y")
else:
  print("Y is larger than X")
```

```python
# Problem 3
user_x = int(input("Please enter integer for x: "))
user_y = int(input("Please enter integer for y: "))
if user_x > user_y:
  print("X is larger than Y")
elif user_x < user_y:
  print("Y is larger than X")
else:
  print("X and Y are equal")
```

```python
# Problem 4
num = 1
while num <= 100:
  print(num)
  num += 1
num = 100
while num >= 1:
  print(num)
  num -= 1
```
